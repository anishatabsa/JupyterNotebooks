```python
class MinWindow:
    def __init__(self, string_array: list):
        self.main_string : str = string_array[0]
        self.subset : str = string_array[1]
        self.main_string_length : int = len(self.main_string)
        self.sub_string_length : int = len(self.subset)
        self.left : int = 0
        self.right : int = 0
        self.unique_chars : list = list(set(self.subset))
        self.min_sub_strings : list = []

    def move_right_pointer(self, position=0):
        i = position
        str_len = i + self.sub_string_length
        print(f'str_len: {str_len}')

        while str_len <= self.main_string_length:
            sub = self.main_string[i : i + str_len]
            sub = [''.join(c) for c in sub]
            print(f'sub: {sub}')
            if self.compare(sub):
                # subs = ''.join(c for c in sub_copy)
                # self.save_min_string(subs)
                self.right = str_len
                break
            else:
                str_len += 1
                print(f'str_len: {str_len}')

    def move_left_pointer(self, position=0):
        i = position
        str_len = i + self.right
        print(f'str_len: {str_len}')
        sub = self.main_string[i : str_len]
        sub_copy = sub

        while str_len <= self.main_string_length:
            sub = self.main_string[i : str_len]
            sub = [''.join(c) for c in sub]
            print(f'sub: {sub}')
            if self.compare(sub):
                i += 1
                sub_copy = sub.copy()
            else:
                subs = ''.join(c for c in sub_copy)
                self.save_min_string(subs)
                self.left = i

    def compare(self, sub_string: list) -> bool:
        chars_present = []
        for char in self.subset:
            if char in sub_string:
                sub_string.pop(sub_string.index(char))
                print(f'sub_pop: {sub_string}')
                chars_present.append(True)
            else:
                chars_present.append(False)
        print(f'chars_present: {chars_present}')
        if all(chars_present):
            return True
        else:
            return False

    def save_min_string(self, sub_string):
        self.min_sub_strings.append(sub_string)
        print(f'search_strings: {self.min_sub_strings}')

    def get_minimum_substring(self) -> str:
        if self.main_string == self.subset:
            return self.subset
        elif self.main_string_length < self.sub_string_length:
            return "Sub string > String"
        elif not self.main_string_length == self.sub_string_length:
            while self.right <= self.main_string_length:
                self.move_right_pointer()
                self.move_left_pointer()
                self.min_sub_strings.sort(key=lambda s: len(s))

        return self.min_sub_strings[0]



#
#
# def get_search_strings(string: str, sub_str: str):
#     search_strings = []
#
#     i = 0
#     str_len = i + len(sub_str)
#     print(f'str_len: {str_len}')
#     while True:
#         chars_present = []
#         sub = string[i : i + str_len]
#         sub = [''.join(c) for c in sub]
#         subt = sub.copy()
#         print(f'sub: {sub}')
#         for char in sub_str:
#             if char in sub:
#                 sub.pop(sub.index(char))
#                 print(f'sub_pop: {sub}')
#                 chars_present.append(True)
#             else:
#                 chars_present.append(False)
#
#         print(f'chars_present: {chars_present}')
#         if all(chars_present):
#             subs = ''.join(c for c in subt)
#             search_strings.append(subs)
#             print(f'search_strings: {search_strings}')
#             i += 1
#             str_len = i + len(sub_str)
#         else:
#             str_len += 1
#
#         print(f'str_len: {str_len}')
#         if str_len > len(string):
#             break
#
#     search_string_list = [(ind, s, len(s)) for ind, s in enumerate(search_strings)]
#     print(f'search_string_list: {search_string_list}')
#     sorted(search_string_list, key=lambda x: x[2])
#     print(f'search_string_list_sorted: {search_string_list}')
#     return search_string_list
#
#
# def minSubstring(strArr: list):
#     string : str = strArr[0]
#     sub_string : str = strArr[1]
#     string_length = len(string)
#     sub_string_length = len(sub_string)
#
#     if string == sub_string:
#         return sub_string
#     elif len(sub_string) > len(string):
#         return "Sub string > String"
#     elif not sub_string_length == string_length:
#         searchString = get_search_strings(string, sub_string)[0][1]
#         print(searchString)
#
```

```python
def LongestWord(sen):
    long_words = ['']
    length = 0
    sen = sen.split(' ')
    print(f'sen: {sen}')

    for word in sen:
        word = word.replace('!', '').replace('&', '')
        print(f'word: {word}')
        word_length = len(word)
        print(f'word_length: {word_length}')

        if word_length > length:
            length = word_length
            print(f'length: {length}')
            long_words.pop(0)
            print(f'long_words: {long_words}')
            long_words.append(word)
            print(f'long_words: {long_words}')

    return long_words[0]
```
